```python
chance = [59.26164, 2.608696, 52.41051, 9.42029, 4.130435, 1.304348,
          34.95169, 17.16222, 17.19074, 66.14849, 22.80124, 13.33333,
          56.5498, 7.539683, 31.3587, 56.456, 72.42067, 11.96817, 18.26087,
          28.91615, 31.17754, 28.04348, 7.125604, 22, 42.65192, 69.8913,
          18.69565, 9.953565, 22.89795, 18.17391, 20.65217, 7.757993,
          71.31926, 65.823, 11.09903, 11.16667, 12.54083, 63.61046, 29.05185]
for i in range(len(chance)) :
    chance[i] = chance[i] / 100
```

```python
def func(m,k):
    if m < k :
        return 0;
    if k < 0 :
        return 0
    else :
        if m == 0 :
            return 1;
        else :
            return (chance[m-1]*func(m-1,k-1)+(1-chance[m-1])*func(m-1,k))
```
