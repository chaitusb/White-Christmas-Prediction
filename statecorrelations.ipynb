```python
import pandas as pd
import numpy as np
import csv
import datetime
from matplotlib import pyplot as plt

cities = ['ALBANY', 'ANNAPOLIS', 'ATLANTA', 'AUGUSTA', 'AUSTIN', 'BATON ROUGE', 'BISMARCK',
 'BOISE', 'BOSTON', 'CARSON CITY', 'CHARLESTON', 'CHEYENNE', 'COLUMBIA',
 'COLUMBUS', 'CONCORD', 'DENVER', 'DES MOINES', 'DOVER', 'FRANKFORT',
 'HARRISBURG', 'HARTFORD', 'HELENA', 'HONOLULU', 'INDIANAPOLIS', 'JACKSON',
 'JEFFERSON CITY', 'JUNEAU', 'LANSING', 'LINCOLN', 'LITTLE ROCK', 'MADISON',
 'MONTGOMERY', 'MONTPELIER', 'NASHVILLE', 'OKLAHOMA CITY', 'OLYMPIA', 'PHOENIX',
 'PIERRE', 'PROVIDENCE', 'RALEIGH', 'RICHMOND', 'SACRAMENTO', 'SAINT PAUL',
 'SALEM', 'SALT LAKE CITY', 'SANTA FE', 'SPRINGFIELD', 'TALLAHASSEE', 'TOPEKA',
 'TRENTON']
rawData = {}
for city in cities:
    cityData = pd.read_csv(city+'.csv',index_col=0, usecols=[0, 'TMIN'], parse_dates=True)
    cityData.columns = [city]
    rawData[city] = cityData

f2 = 'B:\jieminsdata\model\us_state_capitals_coord.csv'
loc = pd.read_csv(f2, usecols=['State','Capital','Lat','Lon'], index_col='Capital')
temp = loc['State'].reset_index()
temp.index = temp['State']
state2capital = temp['Capital'].str.upper().to_dict()
```

```python
data = rawData.values()[0].join(rawData.values()[1:], how='outer')
# Converts data to only include data on christmas
#startDate = data.index[0]
#endDate = data.index[-1]
#if endDate < startDate:
#    endDate = startDate
#xmasIndicies = pd.date_range(start=startDate, end=endDate, freq='A').to_series().shift(periods=-1,freq='6D').index
#data = data.reindex(index=xmasIndicies)
#data = data.sort_index(axis=1)

print data.describe().transpose()
#print data[data['LINCOLN'] == data['LINCOLN'].max()].transpose()
```

```python
corr = data.corr(method='pearson').sort(columns='AUGUSTA',axis=0,ascending=False).transpose().sort(columns='AUGUSTA',axis=0,ascending=False)
#print data.dropna().corr(method='pearson').sort(columns='AUGUSTA',axis=0,ascending=False).transpose().sort(columns='AUGUSTA',axis=0,ascending=False)
#print corr
fig, ax = plt.subplots()
heatmap = ax.pcolor(corr.fillna(value=1, method=None).as_matrix(), cmap=plt.cm.Blues)
plt.show()
print corr.as_matrix()
```

```python
from sklearn import cluster
import cartopy.crs as ccrs
import cartopy.io.shapereader as shpreader

clusters = 5
km = cluster.KMeans(n_clusters=clusters)
c = km.fit_predict(data.dropna().transpose())
for i in range(clusters):
    print 'Cluster: '+str(i)
    print data.columns.values[c==i]
print 'Clustering Inertia: '+str(km.inertia_)


capital2cluster = dict([(data.columns.values[i], c[i]) for i in range(len(c))])
cluster2color = dict([(0, 'red'),(1, 'blue'),(2, 'green'),(3,'yellow'),(4,'orange'),(5, 'gray')])
    
f = plt.figure(figsize=(12,12))
ax = plt.axes([0, 0, 1, 1],projection=ccrs.LambertConformal())
ax.set_extent([-125, -66.5, 20, 50], ccrs.Geodetic())
shapename = 'admin_1_states_provinces_lakes_shp'
states_shp = shpreader.natural_earth(resolution='110m',category='cultural', name=shapename)

# to get the effect of having just the states without a map "background"
# turn off the outline and background patches
ax.background_patch.set_visible(False)
ax.outline_patch.set_visible(False)

for state in shpreader.Reader(states_shp).records():
    facecolor = 'white'
    if state.attributes['name'] in state2capital.keys():
        facecolor = cluster2color[capital2cluster[state2capital[state.attributes['name']]]]
    edgecolor = 'black'
    ax.add_geometries([state.geometry], ccrs.PlateCarree(),facecolor=facecolor, edgecolor=edgecolor)

plt.title('K-Means Clustering with '+str(clusters)+' Clusters')
#savefig('AvgAbsErr-datamap.png', bbox_inches='tight')
plt.show()
```
