```python
import pandas as pd
import numpy as np
import csv
import datetime

cities = ['ALBANY', 'ANNAPOLIS', 'ATLANTA', 'AUGUSTA', 'AUSTIN', 'BATON ROUGE', 'BISMARCK',
 'BOISE', 'BOSTON', 'CARSON CITY', 'CHARLESTON', 'CHEYENNE', 'COLUMBIA',
 'COLUMBUS', 'CONCORD', 'DENVER', 'DES MOINES', 'DOVER', 'FRANKFORT',
 'HARRISBURG', 'HARTFORD', 'HELENA', 'HONOLULU', 'INDIANAPOLIS', 'JACKSON',
 'JEFFERSON CITY', 'JUNEAU', 'LANSING', 'LINCOLN', 'LITTLE ROCK', 'MADISON',
 'MONTGOMERY', 'MONTPELIER', 'NASHVILLE', 'OKLAHOMA CITY', 'OLYMPIA', 'PHOENIX',
 'PIERRE', 'PROVIDENCE', 'RALEIGH', 'RICHMOND', 'SACRAMENTO', 'SAINT PAUL',
 'SALEM', 'SALT LAKE CITY', 'SANTA FE', 'SPRINGFIELD', 'TALLAHASSEE', 'TOPEKA',
 'TRENTON']
rawData = {}
for city in cities:
    cityData = pd.read_csv(city+'.csv',index_col=0, usecols=[0, 'SNWD'], parse_dates=True)
    cityData.columns = [city]
    rawData[city] = cityData
```

```python
data = rawData.values()[0].join(rawData.values()[1:], how='outer')
startDate = data.index[0]
endDate = data.index[-1]
if endDate < startDate:
    endDate = startDate
xmasIndicies = pd.date_range(start=startDate, end=endDate, freq='A').to_series().shift(periods=-1,freq='6D').index
data = data.reindex(index=xmasIndicies)
data = data.sort_index(axis=1)
print data
```

```python
f = 'B:\jieminsdata\model\\baseline_ext_eval_input_diff_sorted.csv'
actualData = pd.read_csv(f, usecols=['key', 'actual'], parse_dates = False)
actualData.columns = ['City','Outcome']
key = actualData.loc[:,'City'].str.split(pat=',',n=1)
actualData.loc[:,'City'] = key.str[0].str.upper()
actualData = actualData.join(pd.Series(data=pd.to_datetime(key.str[1]), name='Date'), how='left')
actualData.loc[actualData.loc[:,'City']=='HELANA', 'City'] = 'HELENA'
print actualData
```

```python
#NOTE: Snowdepth is in mm. There are 25.4 millimeters in an inch.
#Marks days with snowdepth >= 1 inch as 1 and days with snowdepth < 1 inch as 0
data[data < 25.4] = 0
data[data >= 25.4] = 1
print data
```
